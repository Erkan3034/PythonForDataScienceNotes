## args(arguments)  - kwarg( keyword arguments)

❗❗``*args`` → Fonksiyona ***istediğin kadar*** konumsal parametre göndermeni sağlar.  
Fonksiyon içinde tuple olarak tutulur.

```python
def args_example(*args):
    return sum(args)

args_example(10,20,30,40,50,60,70)
```

```python
def topla(*args):
    print(args)

topla(1, 2, 3, 4)
```

____________

❗❗``**kwargs`` → Fonksiyona istediğin kadar ***anahtar = değer (keyword argument)*** göndermeni sağlar.  
Fonksiyon içinde ***``dict``*** olarak tutulur.

```python
def bilgi(**kwargs):
    print(kwargs)

bilgi(ad="Erkan", yas=20)
```

```python
def detaylar(**kwargs):
    for key, value in kwargs.items():
        print(key, "→", value)

detaylar(marka="BMW", model="M4", yil=2025)
```

```python
def ornek(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

ornek(1, 2, 3, ad="Erkan", seviye="Junior")
```
