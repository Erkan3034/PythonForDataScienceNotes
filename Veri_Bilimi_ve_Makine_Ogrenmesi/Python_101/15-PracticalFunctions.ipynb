## zip()  

``zip`` â†’ iki (veya daha fazla) listeyi eleman eleman ***birleÅŸtirir.***

```python
a = [1, 2, 3]
b = ["a", "b", "c"]

sonuc = list(zip(a, b))
print(sonuc)
```

EÅŸleÅŸme mantÄ±ÄŸÄ±  

Zip her zaman aynÄ± indexli elemanlarÄ± bir araya getirir.  

-
a: 1   2   3  
b: a   b   c  
--|--|--|  
zip â†’ (1,a), (2,b), (3,c)
- 

```python
#Zip en kÄ±sa liste kadar eÅŸleÅŸtirme yapar.

a = [1, 2, 3]
b = ["a", "b"]

sonuc = list(zip(a, b))
print(sonuc)
```

```python
# UNZIP
pairs = [(1, "a"), (2, "b"), (3, "c")]

x, y = zip(*pairs) # unzip iÃ§in * kullanÄ±lÄ±r
print(x)  # (1, 2, 3)
print(y)  # ("a", "b", "c")
```
