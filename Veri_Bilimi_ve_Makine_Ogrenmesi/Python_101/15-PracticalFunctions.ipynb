## map()

``map``: Bir listenin her elemanÄ±na iÅŸlem uygular.

```python
sayilar = [10,20,30,40,50]

print(list(map(lambda x : x*2 , sayilar)))
```

```python
def controlStr(s):
    return "Erkan" in s


myStrList= ["Erkan", "Serkan"]

print(list(map(controlStr,myStrList)))
```

## filter  

``filter`` : ÅžARTI saÄŸlayan elemanlarÄ± alÄ±r, diÄŸerlerini ele.

```python
sayilar = [1, 2, 3, 4, 5, 6]
ciftler = list(filter(lambda x: x % 2 == 0, sayilar))
print(ciftler)
```

## zip()  

``zip`` â†’ iki (veya daha fazla) listeyi eleman eleman ***birleÅŸtirir.***

```python
a = [1, 2, 3]
b = ["a", "b", "c"]

sonuc = list(zip(a, b))
print(sonuc)
```

EÅŸleÅŸme mantÄ±ÄŸÄ±  

Zip her zaman aynÄ± indexli elemanlarÄ± bir araya getirir.  

-
a: 1   2   3  
b: a   b   c  
--|--|--|  
zip â†’ (1,a), (2,b), (3,c)
- 

```python
#Zip en kÄ±sa liste kadar eÅŸleÅŸtirme yapar.

a = [1, 2, 3]
b = ["a", "b"]

sonuc = list(zip(a, b))
print(sonuc)
```

```python
# UNZIP
pairs = [(1, "a"), (2, "b"), (3, "c")]

x, y = zip(*pairs) # unzip iÃ§in * kullanÄ±lÄ±r
print(x)  # (1, 2, 3)
print(y)  # ("a", "b", "c")
```

```python
### SCOPE (LEGB)

print("---- GLOBAL SCOPE ----")
a = 50  # global deÄŸiÅŸken




print("Global a:", a)

print("\n---- LOCAL SCOPE ----")
def local_example():
    x = 10  # local
    print("Fonksiyon iÃ§indeki x:", x)

local_example()
# print(x)  # HATA olur (local dÄ±ÅŸarÄ±dan gÃ¶rÃ¼nmez)




print("\n---- ENCLOSING SCOPE ----")
def outer():
    y = 20  # enclosing deÄŸiÅŸken

    def inner():
        print("inner iÃ§inde y:", y)  # enclosing'e eriÅŸir

    inner()

outer()





print("\n---- NONLOCAL Ã–RNEK ----")
def outer2():
    z = 30

    def inner2():
        nonlocal z
        z = 99
    inner2()
    print("nonlocal sonrasÄ± z:", z)

outer2()




print("\n---- GLOBAL DEÄžÄ°ÅžTÄ°RME Ã–RNEÄžÄ° ----")
b = 5

def change_global():
    global b
    b = 100

change_global()
print("Global b deÄŸiÅŸti:", b)




print("\n---- LEGB TEST ----")
a = 50  # global

def f1():
    a = 20  # enclosing
    def f2():
        print("f2 iÃ§inde gÃ¶rÃ¼nen a:", a)  # enclosing'a bakacak
    f2()

f1()





print("\n---- LOCAL SHADOWING (DEÄžÄ°ÅžTÄ°RMEYEN) ----")
c = 7

def shadow():
    c = 999   # bu global c'yi deÄŸiÅŸtirmez (yeni local oluÅŸturur)
    print("Fonksiyon iÃ§indeki (local) c:", c)

shadow()
print("Global c aynÄ± kaldÄ±:", c)
```
