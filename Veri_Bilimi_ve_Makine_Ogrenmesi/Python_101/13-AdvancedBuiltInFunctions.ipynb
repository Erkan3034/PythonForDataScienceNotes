## enumerate => index ekler

```python
meyveler = ["elma", "muz", "portakal"]

for indeks, meyve in enumerate(meyveler):
    print(f"{indeks}: {meyve}")
```

```python
for (idx,val) in enumerate(list(range(1,10))):
    print(idx, ":",val)
```

```python
for myList in enumerate(list(range(1,10))):
    print(myList)
```

## zip()

**zip(), birden fazla listeyi (veya iterable’ı) elemanlarına göre yan yana eşleştiren bir fonksiyon.**
*Yani liste 1’in 0. elemanı → liste 2’nin 0. elemanı*
**liste 1’in 1. elemanı → liste 2’nin 1. elemanı**
**… diye gider.**
___

```python
isimler = ["Erkan", "Umut", "Berk"]
yaslar = [20, 21, 19]

sonuc = list(zip(isimler, yaslar))
print(sonuc)
```

```python
a = [1, 2, 3]
b = ["a", "b", "c"]
c = [True, False, True]

print(list(zip(a, b, c)))
```

```python
for isim, yas in zip(isimler, yaslar):
    print(isim, yas)
```
