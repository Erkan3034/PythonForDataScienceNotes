Nesnenin ilk özellikleri burada tanımlanır.

___

``Method Nedir?``

Class içinde tanımlanan fonksiyonlardır.

```python
class Araba:
    # nesne oluştururken çalışan constructor metot
    def __init__(self,marka,model,renk,hiz=0): 
        self.marka = marka
        self.model = model
        self.renk  = renk
        self.hiz   = hiz

    def hizlan(self,miktar):
        self.hiz+=miktar
        print(f"\n{self.marka}  {self.model} hızlandı! Yeni hizi : {self.hiz}\n")

    def yavasla(self,miktar):
        self.hiz-=miktar
        print(f"\n{self.marka}  {self.model}  yavaşladı! Yeni hız: {self.hiz}\n")

    def bilgi_goster(self):
        print(f"\n*********Araç Bilgileri**********\n\n Araç Markası: {self.marka}\n Araç Model  : {self.model}\n Araç Renk   : {self.renk}\n Araç Hız    : {self.hiz}")


araba = Araba("Dodge", "Challenger" , "Siyah" , 280)

araba2 = Araba("Audi" , "A5" , "Siyah" , 290)


araba.hizlan(50)

araba.bilgi_goster()

araba2.bilgi_goster()

araba2.hizlan(70)
```

# Inheritance (Kalıtım)

**Bir sınıf başka bir sınıftan özellik ve metot miras alır.**

`` Bir Temel sınıf => bir Alt sınıf``  

Gerçek hayat benzetmesi:

“Araba” diye bir sınıfı var.

“Elektrikli Araba” özel bir araba türüdür. Arabanın tüm özelliklerini alır, üstüne yenilerini ekler.

```python
class Arac:
    def __init__(self, marka,model):
        self.marka = marka
        self.model = model

    def bilgi(self):
        print(f"{self.marka} {self.model} \n")


#Arac sınıfında Miras alıyoruz
class Araba(Arac): 
    
    def __init__(self, marka, model, renk):
        super().__init__(marka, model)
        self.renk = renk
        


    def bilgi(self):
        super().bilgi() #üst sınıfın fonksiyonu.
        print(f"Renk: {self.renk} \n")



#Araba sınıfından miras alıyoruz
class Tır(Araba): 
    def __init__(self,marka,model,renk,teker_sayısı):
        super().__init__(marka,model,renk)
        self.teker_sayısı = teker_sayısı

    def bilgi(self):
        super().bilgi() # araba sınıfının fonksiyonu
        print(f"Teker Sayısı: {self.teker_sayısı}")


a1 = Araba("Tesla" , "Model 3" , "Kırmızı")

a1.bilgi()
print("__________________________\n")

t1 = Tır("SCANIA" ,"Kırk Ayak" , "Gri" , 16)
t1.bilgi()
```

# Polymorphism (Çok Biçimlilik)

***Aynı isimli metodun farklı sınıflarda farklı şekilde çalışmasıdır***

✔ Örnek:

Her hayvan “ses çıkarır” ama çıkarılan ses farklıdır.

```python
class Hayvan:
    def ses(self):
        print("HAYVAN: Bir hayvan sesi geliyor")

class Kopek:
    def ses(self):
        print("KÖPEK: Hav hav")


class Kedi:
    def ses(self):
        print("KEDİ: Miyauuuvvv")


hayvanlar = [Hayvan() ,Kedi() , Kopek()]

for h in hayvanlar:
    h.ses()
```

# Encapsulation (Kapsülleme / Gizleme)

Veriyi koruma olayı.

Python'da ``_ ``ve ``__`` ile yapılır.

***Örneğin: Bankadaki bakiye dışarıdan direkt değiştirilemez, methodlarla kontrol edilir.***

```python
class Banka:
    def __init__(self):
        self.__bakiye = 0   # private değişken

    def para_yatir(self, miktar):
        self.__bakiye += miktar

    def para_cek(self, miktar):
        if miktar <= self.__bakiye:
            self.__bakiye -= miktar
        else:
            print("Yetersiz bakiye!")

    def bakiye_goster(self):
        print(f"Bakiye: {self.__bakiye}")

b = Banka()
b.para_yatir(100)
b.para_cek(20)
b.bakiye_goster()
```

# Abstraction (Soyutlama)

Gereksiz detayları gizleyip kullanıcıya sade bir yapı sunmak.

✔ Örnek:

Arabayı sürerken motorun içini bilmek zorunda değilsin → gaz pedalı yeterli.

```python
from abc import ABC, abstractmethod # soyut sınıf oluşturmak için kullanıyoruz

class Odeme(ABC):
    @abstractmethod
    def odeme_yap(self, miktar):
        pass

class KrediKart(Odeme):
    def odeme_yap(self, miktar):
        print(f"{miktar} TL kredi kartı ile ödendi.")

class Nakit(Odeme):
    def odeme_yap(self, miktar):
        print(f"{miktar} TL nakit ödendi.")

o1 = Nakit()
o1.odeme_yap(100)

o2 = KrediKart()
o2.odeme_yap(150)
```

```python
# ================================
#    OOP E-Ticaret Sepet Sistemi
# ================================

DISCOUNT_RATE = 0.10   # %10 indirim

# -------------------------------
# Temel Ürün Sınıfı (Base Class)
# -------------------------------
class Product:
    def __init__(self, name, price):
        self.name = name
        self.__price = price  # encapsulation

    def get_price(self):
        """İndirimsiz fiyat"""
        return self.__price

    def get_discounted_price(self):
        """Her ürün %10 indirimli fiyat"""
        return self.__price * (1 - DISCOUNT_RATE)

    def get_info(self):
        return f"{self.name} - {self.get_discounted_price()}₺"


# -------------------------------
# Elektronik Ürün
# -------------------------------
class Electronic(Product):
    def __init__(self, name, price, warranty):
        super().__init__(name, price)
        self.warranty = warranty

    def get_info(self):
        return f"{self.name} - {self.get_discounted_price()}₺ (Garanti: {self.warranty} yıl)"


# -------------------------------
# Giyim Ürün
# -------------------------------
class Clothing(Product):
    def __init__(self, name, price, size):
        super().__init__(name, price)
        self.size = size

    def get_info(self):
        return f"{self.name} - {self.get_discounted_price()}₺ (Beden: {self.size})"


# -------------------------------
# Kitap Ürün
# -------------------------------
class Book(Product):
    def __init__(self, name, price, author):
        super().__init__(name, price)
        self.author = author

    def get_info(self):
        return f"{self.name} - {self.get_discounted_price()}₺ (Yazar: {self.author})"


# -------------------------------
# Sepet Sınıfı
# -------------------------------
class Cart:
    def __init__(self):
        self.items = []

    def add(self, product):
        self.items.append(product)
        print(f"{product.name} sepete eklendi!")

    def total(self):
        return sum(item.get_discounted_price() for item in self.items)

    def list_items(self):
        print("\n--- Sepetteki Ürünler ---")
        for item in self.items:
            print(item.get_info())
        print("---------------------------")
        print(f"Toplam (İndirimli): {self.total()}₺\n")


# -------------------------------
# Örnek Kullanım
# -------------------------------
if __name__ == "__main__":
    iphone = Electronic("iPhone 14", 45000, 2)
    tshirt = Clothing("Tişört", 300, "L")
    book = Book("Python 101", 250, "Guido")

    sepet = Cart()
    sepet.add(iphone)
    sepet.add(tshirt)
    sepet.add(book)

    sepet.list_items()
```
