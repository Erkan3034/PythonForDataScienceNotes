```python
from typing import List, Dict, Tuple, Optional, Union, Callable

 
#  Değişkenlerde Type Annotation

yas: int = 20
isim: str = "Erkan"
fiyat: float = 99.9
aktif: bool = True

```

```python
#  Fonksiyonlarda Type Annotation

def kare(x: int) > int:
    return x * x

def mesaj_ver(isim: str) > str:
    return f"Merhaba {isim}"

def bolme(a: float, b: float) > Optional[float]:
    if b == 0:
        return None
    return a / b


```

```python


#  List, Dict, Tuple Annotation
 
sayilar: List[int] = [1, 2, 3, 4]
urun: Dict[str, float] = {"fiyat": 100.0, "kdv": 0.18}
konum: Tuple[int, int] = (10, 20)


```

```python

#  Union (birden fazla tip olabilir)

def yazdir(x: Union[int, str]) > None:
    print(x)

# Python 3.10+ alternatifi:
def yazdir2(x: int | str) > None:
    print(x)

```

```python

 
#  *args ve **kwargs Type Annotation
 
def topla(*sayilar: int) > int:
    return sum(sayilar)

def ayarlar(**kwargs: str) > None:
    print(kwargs)

```

```python

 
# Lambda + Callable
 
kare_lambda: Callable[[int], int] = lambda x: x * x

```

```python


# OOP + Type Annotation
 
class Urun:
    isim: str
    fiyat: float
    indirim_orani: float

    def __init__(self, isim: str, fiyat: float, indirim_orani: float = 0.10):
        self.isim = isim
        self.fiyat = fiyat
        self.indirim_orani = indirim_orani

    def indirimli_fiyat(self) >float :
        return self.fiyat * (1  self.indirim_orani)

    def __str__(self) > str:
        return f"{self.isim}  {self.indirimli_fiyat():.2f} TL"



#  Kullanım Örnekleri (çıktı görmek için)

print(kare(5))
print(mesaj_ver("Erkan"))
print(bolme(10, 2))
print(bolme(10, 0))

yazdir(10)
yazdir("Python")

print(topla(1, 2, 3, 4))
ayarlar(tema="dark", dil="tr")

urun1 = Urun("Laptop", 30000)
urun2 = Urun("Mouse", 500, 0.20)

print(urun1)
print(urun2)
print(kare_lambda(6))
```
