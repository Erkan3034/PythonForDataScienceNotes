``init Nedir?``

__init__ = constructor = “nesne oluşturulurken çalışan özel fonksiyon”.` 

Nesnenin ilk özellikleri burada tanımlanır.

___

``Method Nedir?``

Class içinde tanımlanan fonksiyonlardır.

```python
class Araba:
    # nesne oluştururken çalışan constructor metot
    def __init__(self,marka,model,renk,hiz=0): 
        self.marka = marka
        self.model = model
        self.renk  = renk
        self.hiz   = hiz

    def hizlan(self,miktar):
        self.hiz+=miktar
        print(f"\n{self.marka}  {self.model} hızlandı! Yeni hizi : {self.hiz}\n")

    def yavasla(self,miktar):
        self.hiz-=miktar
        print(f"\n{self.marka}  {self.model}  yavaşladı! Yeni hız: {self.hiz}\n")

    def bilgi_goster(self):
        print(f"\n*********Araç Bilgileri**********\n\n Araç Markası: {self.marka}\n Araç Model  : {self.model}\n Araç Renk   : {self.renk}\n Araç Hız    : {self.hiz}")


araba = Araba("Dodge", "Challenger" , "Siyah" , 280)

araba2 = Araba("Audi" , "A5" , "Siyah" , 290)


araba.hizlan(50)

araba.bilgi_goster()

araba2.bilgi_goster()

araba2.hizlan(70)
```
