# Veri Tipi Dönüştürme (Type Casting)

Python’da bazı veri tiplerini diğerine dönüştürebiliriz.

```python
x ="25"
```

```python
type(x)
```

```python
x = int(x)
type(x)
```

```python
x = float(x)
type(x)
```

```python
y = 30
type(y)
```

```python
z = str(y)
type(z)
```

### List -> Tuple

________________________________________

```python
liste = [10,20,30,40]
```

```python
tpl = tuple(liste)
```

```python
liste = list(tpl) # tuple->list 
```

```python
liste
```

```python
liste.append(1)
```
