## LOOPS
_____

```python
for i in range(1, 10):
    print([x * i for x in range(1, 11)])  # multiples from 1×i to 10×i
```

```python
myTuple = (10,20,30,40)
for x in myTuple:
    print(x)
```

```python
# Simple tuple unpacking
point = (3, 7)
x, y = point
print(x)  # 3
print(y)  # 7
```

```python
for i in range (50):
    if not i == 40: # if i == 40, skip it(don't write 40) 
        print(i)
    else:
        break
```

```python
for i in range(50):
    if i == 40:
        continue
    print(i)
```
