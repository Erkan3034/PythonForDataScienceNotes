# Dictionary (Sözlük)

```python
# key - value pairing.
```

```python
fruit_list = {
    "Banana" : 150,
    "Apple" : 100
}
```

```python
type(fruit_list)
```

```python
print(fruit_list["Banana"])
```

```python
# DEĞER DEĞİŞTİRİLEBİLİR
fruit_list['Banana'] = 250
```

```python
fruit_list['Banana']
```

```python
#YENI PAIRLER EKLENEBİLİR
fruit_list['Orange'] = 180
```

```python
fruit_list['Orange']
```

```python
# BIR KAYIT SILMEK:
del fruit_list['Banana']
```

```python
# FOR ILE DICT UZERINDE DOLASMAK
for i in fruit_list:
    print(fruit_list[i])
```

```python
fruit_list.keys() # sadefe anahtarları al
```

```python
fruit_list.values() # sadece değerleri al
```

```python
type(fruit_list.values())
```

```python
# FOR İLE KALROİLERİN ORTALAMASI:
values = list(fruit_list.values())
type(values) # dict_values -> list

top = 0
for i in values:
    top+=i
    
ort = top /len(values)
print(ort)
```
