# Slicing nedir ?

Bir liste, string veya tuple gibi dizilerin içinden belli bir parçayı kesip alma işlemidir.

``Slicing = “dilimleme”.``

### Temel Slicing Yapısı

`
liste[start : stop : step]
`


*start* → nereden başlasın

*stop* → nereye kadar gitsin (stop dahil değildir)

*step* → kaçar kaçar ilerlesin

```python
barcode = "ABCD123465454" # barkodun ilk 3 harfinden sonra gelen ilk  hariç harf vebar rakamlardan ilk 5'ini alalım

sliced_barcode = barcode[3:9] 
print(sliced_barcode)
```

```python
print(barcode[::-1]) #barkodu tersten yazdırma
```

```python
print(barcode[-5:]) #sondan ilk 5 eleman
```

```python
#Sondan geriye doğru 5 karakter
print(barcode[:-6:-1])
```
