```python
def simple_counter(limit=5):
        """Print numbers 0..limit-1 using a while loop."""
        i = 0
        while i < limit:
                print(i)
                i += 1  # important: update the loop variable

simple_counter()
```

```python

def find_first_even(nums):
        """
        Scan a sequence for the first even number.
        Demonstrates break and while-else: else runs if no break occurred.
        """
        i = 0
        while i < len(nums):
                if nums[i] % 2 == 0:
                        print("Found first even:", nums[i])
                        break
                i += 1
        else:
                print("No even found")

find_first_even([1,3,5,8,4,9,12])
```

```python

def remove_until_empty(stack):
        """Consume a list as a stack until it's empty (truthiness check)."""
        while stack:  # while stack is non-empty
                item = stack.pop()
                print("Popped", item)

remove_until_empty([1,2,3,4,5])
```

```python
def skip_odds(limit=10):
        """Show continue: skip odd numbers and print only evens."""
        i = 0
        while i < limit:
                i += 1
                if i % 2 == 1:
                        continue
                print("even", i)

skip_odds()
```

```python
def infinite_example(max_iter=3):
        """
        Example of an infinite loop that is intentionally broken out of.
        Useful when the stopping condition is detected inside the loop.
        """
        count = 0
        while True:
                print("loop", count)
                count += 1
                if count >= max_iter:
                        break
infinite_example()
```
